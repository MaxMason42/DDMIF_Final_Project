```python
def read_and_get_returns(file_paths):
    dfs = [pd.read_csv(file) for file in file_paths]
    df = pd.concat(dfs, ignore_index=True)

    df['time'] = pd.to_datetime(df['time'])

    num_identifiers = None

    if not num_identifiers:
        num_identifiers = len(df.dropna()["identifier"].unique())
    
    avg_returns = df.dropna().groupby('time')["captured_returns"].sum()/num_identifiers

    return (1 + avg_returns).cumprod() - 1
```

```python
momentum = read_and_get_returns(momentum_file_paths)
long = read_and_get_returns(long_only_file_paths)
lstm_returns = read_and_get_returns(lstm_file_paths)
TFT_returns = read_and_get_returns(TFT_file_paths)
TFT_CPD_returns = read_and_get_returns(TFT_CPD_file_paths)
TFT_extended = read_and_get_returns(TFT_extended_file_paths)
TFT_CPD_extended = read_and_get_returns(TFT_CPD_extended_file_paths)
TFT_heads = read_and_get_returns(TFT_heads_file_paths)
TFT_CPD_heads = read_and_get_returns(TFT_CPD_heads_file_paths)
```

```python
plt.figure(figsize=(12, 6))

plt.plot(long.index, long, label="Long-Only")
plt.plot(momentum.index, momentum, label="Momentum")
plt.plot(lstm_returns.index, lstm_returns, label="LSTM")
plt.plot(TFT_returns.index, TFT_returns, label="TFT")
plt.plot(TFT_CPD_returns.index, TFT_CPD_returns, label="TFT-CPD")
plt.plot(TFT_CPD_heads.index, TFT_CPD_heads, label = "TFT-CPD-6")

plt.xlabel('Date')
plt.ylabel("Cummulative Return")
plt.title("Strategy Returns")

plt.gca().xaxis.set_major_locator(mdates.YearLocator()) 
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

plt.legend()
plt.show()
```
